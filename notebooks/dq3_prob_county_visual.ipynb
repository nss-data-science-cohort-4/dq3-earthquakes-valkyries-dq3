```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
%matplotlib inline
```

```python
prob_county = pd.read_csv('../data/tn_shake_geo.csv')
prob_county.head()
```

```python
prob_county.describe()
```

```python
prob_county.groupby('NAME')['PctProb100yrs'].mean().sort_values(ascending = False)
```

```python
fig, ax = plt.subplots(figsize = (8,6))
sns.boxplot(data = prob_county, x = 'NAME', y = 'PctProb100yrs');
```

```python
prob_county.groupby('NAME')['PctProb100yrs'].mean().sort_values(ascending = True).plot(kind = 'bar', figsize=(10,6))
plt.title('TN Counties Probability of Ground Shaking over next 100 Yrs')
plt.ylabel('Probability')
# internet_df.groupby('Year')['Internet_Users_Pct'].mean().plot(kind = 'bar', figsize=(15,8))
# plt.title('Average Percent of Internet Users by Year, All Countries')
# plt.ylabel('Internet_Users_Pct');
```
