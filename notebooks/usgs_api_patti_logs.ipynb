Apply logarithm on usgs_tn 'mag' (magnitude) to determine if it shows relationship to power law distribution.

```python
# remove '-inf' causing errors, replace with 'nan'
usgs_tn['log_mag'] = usgs_tn['log_mag'].replace(-np.inf, np.nan)
```

Result: neither histogram nor qq-plot appears to indicate that applying log transformation shows that every unit increase change in magnitude reflects a lower occurence.  This is because applying a log transformation causes the data to follow a more normal distribution.  

The original histogram of the usgs_tn['mag'] is the best reflection of the distribution described by Nate Silver, after the median of 1.6 magnitude.  

The ramp-up on the smaller magnitudes (which conflicts with Nate Silver's analysis) may be due to an increased monitoring over time.
